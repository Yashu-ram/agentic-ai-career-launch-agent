# Define State Schema

The state schema is the shared memory used across all nodes in the LangGraph workflow. Each node reads data from the state, updates it, and passes the updated state to the next node.

## State Variables

| State Field | Description |
|---|---|
| `query` | Stores the user input or question provided to the workflow |
| `retrieved_docs` | Stores the information retrieved from the retrieval step |
| `draft_answer` | Stores the generated draft response created during reasoning |
| `safety_flag` | Indicates whether the query is `safe` or `risky` |
| `approved` | Stores the human reviewer approval status (`True` or `False`) |

---

# State Schema Code

```python
from typing import TypedDict


class AgentState(TypedDict):
    query: str
    retrieved_docs: str
    draft_answer: str
    safety_flag: str
    approved: bool

# Define Edges with Conditional Branching at `safety_node`

In LangGraph, edges define the flow between nodes. A conditional branch allows the workflow to take different paths based on the current state.

In this workflow:
- If the query is safe → move directly to `answer_node`
- If the query is risky → move to `review_node` for human approval

---

# Workflow Structure

```text id="kpkbwy"
retrieve
   ↓
reason
   ↓
safety
 /      \
safe    risky
 |         |
answer   review
            ↓
         answer


# Compare This Workflow to a Day 7 Agent Loop

## Day 7 Agent Loop

A simple agent loop follows a repeated cycle:

```python
while True:
    think()
    act()
    observe()
```

### Characteristics
- Linear execution flow
- Simple reasoning cycle
- Minimal state management
- Best for lightweight workflows
- Easier to build for small tasks

### Best Use Cases
- Basic chatbot
- Simple tool calling
- Small automation tasks
- Single-step reasoning systems

---

# LangGraph Workflow

The LangGraph workflow uses explicit nodes, edges, and shared state.

```text
Input
  ↓
Retrieve
  ↓
Reason
  ↓
Safety Check
 /          \
Safe       Risky
 |            |
Answer    Human Review
              ↓
           Answer
```

### Characteristics
- Stateful execution
- Explicit workflow structure
- Conditional branching
- Human-in-the-loop support
- Better orchestration and debugging

### Best Use Cases
- Multi-step AI systems
- Safety-critical workflows
- Approval-based systems
- Complex enterprise workflows
- Production-grade agent orchestration

---

# Key Differences

| Feature | Day 7 Agent Loop | LangGraph Workflow |
|---|---|---|
| Workflow Type | Linear loop | Graph-based workflow |
| State Management | Minimal | Explicit shared state |
| Branching | Difficult to manage | Built-in conditional routing |
| Human Review | Hard to integrate | Easy with review nodes |
| Debugging | Can become messy | Clear workflow visualization |
| Scalability | Limited | Highly scalable |
| Maintainability | Lower for complex logic | Easier for large workflows |

---

# When the Graph Model is Clearer

LangGraph is better when:
- multiple workflow steps exist
- safety checks are required
- human approvals are needed
- retries or checkpoints are required
- workflows contain branching logic

Example:

```text
if safe → answer
if risky → human review
```

Graphs make this logic easier to visualize and maintain.

---

# When LangGraph is Overkill

LangGraph may be unnecessary when:
- the workflow is very simple
- only one tool call is needed
- there is no branching logic
- no state tracking is required

In such cases, a simple agent loop is easier and faster to implement.

# Final Observation

The Day 7 agent loop is ideal for simple, lightweight agents.

LangGraph becomes more effective when workflows require:

LangGraph improves workflow clarity, maintainability, and production readiness for complex AI systems.



# Observations: When is the Graph Model Clearer vs Overkill?

## When the Graph Model is Clearer

The graph model becomes clearer and more useful when workflows contain:
- multiple execution steps
- conditional branching
- safety checks
- retries
- human approval systems
- explicit state transitions

In this LangGraph workflow, the graph structure clearly shows how the agent behaves:

```text
Input
  ↓
Retrieve
  ↓
Reason
  ↓
Safety Check
 /          \
Safe       Risky
 |            |
Answer    Human Review
              ↓
           Answer
```

This visualization makes it easier to:
- understand workflow execution
- debug failures
- track state changes
- manage complex logic
- scale workflows in production systems

Graphs are especially useful in:
- healthcare systems
- legal AI systems
- financial approval workflows
- enterprise automation
- multi-agent orchestration

---

# When the Graph Model is Overkill

The graph model becomes unnecessary when the workflow is simple and linear.

Examples:
- basic chatbot
- simple question-answering system
- single tool-calling agent
- one-step retrieval system

For simple workflows, using LangGraph can:
- add unnecessary complexity
- increase setup time
- make debugging harder
- create too many files for small logic

In such cases, a simple agent loop is usually better:

```python
while True:
    think()
    act()
    observe()
```

---

# Final Observation

Graphs are best for complex workflows requiring:
- state management
- branching logic
- approvals
- retries
- orchestration

Simple loops are better for:
- lightweight agents
- quick prototypes
- straightforward tasks

The key lesson is:
> Use graphs only when workflow complexity justifies explicit state transitions and branching.

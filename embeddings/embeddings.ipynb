What Are Embeddings?

Embeddings are numerical vector representations of text generated by an embedding model. They convert words, sentences, or documents into lists of numbers that preserve semantic meaning. Instead of understanding text as plain words, AI systems compare these vectors mathematically.

For example:

“Python is a programming language”
“Java is a programming language”

produce embeddings that are closer together because they share a similar meaning related to programming.

Whereas:

“Python is a snake”

produces a different embedding because the context refers to an animal instead of software development.

In this project, the all-MiniLM-L6-v2 embedding model from Hugging Face was used to generate 384-dimensional embeddings for each sentence.

Embeddings are one of the most important components in:

Retrieval-Augmented Generation (RAG)
Semantic search
Recommendation systems
Vector databases
AI agents with memory
Cosine Similarity Observations

Cosine similarity was used to compare how semantically close two embeddings are.

The similarity score ranges generally from:

1.0 → identical meaning
0.7+ → highly related
0.4–0.6 → moderately related
0–0.2 → weakly related
negative values → unrelated or opposite meaning
Observations From the Experiment

The sentences:

“Data analysts use SQL and Power BI”
“Python is widely used in data science”

had one of the highest similarity scores because both belong to the data and analytics domain.

The sentences:

“Doctors diagnose diseases in hospitals”
“Graphic designers create visual content”

had the lowest similarity score because they belong to unrelated professional domains.

The query:

“AI and machine learning”

correctly retrieved:

“Machine learning engineers build AI systems”

with the highest similarity score, demonstrating that embeddings understand semantic meaning instead of exact keywords.

This experiment showed that embeddings can identify contextual relationships between sentences even when the wording is different.

Why Semantic Search Beats Keyword Search

Traditional keyword search relies on exact word matching. It often fails when different words are used to describe similar concepts.

Example:

Query

“AI engineer jobs”

Document

“Machine learning developers build intelligent systems”

Keyword search may fail because the exact words “AI engineer” are not present.

Semantic search succeeds because embeddings understand that:

AI engineers
machine learning developers
intelligent systems

are conceptually related.

Advantages of Semantic Search
Understands meaning instead of exact words
Handles synonyms effectively
Improves document retrieval quality
Better for natural language queries
Essential for modern RAG systems

Semantic search is widely used in:

Google search systems
Netflix recommendations
LinkedIn job matching
AI chatbots and document retrieval systems
Challenges and Limitations

Although embeddings are powerful, several challenges were observed during the experiment.

1. Low Similarity Scores

Some relevant sentences still produced relatively low similarity values. This can happen because:

sentences are too short
context is limited
embedding model is general-purpose
2. Chunking Problems

Large chunks of text may reduce retrieval quality because multiple topics become mixed together. Proper chunking is important in RAG pipelines.

3. Domain-Specific Knowledge

General embedding models may not perform well in highly specialized domains such as:

medicine
law
chemistry
finance

In such cases, domain-specific embedding models may be required.

4. Embeddings Do Not Generate Text

A common misunderstanding is assuming embeddings generate responses. Embeddings only represent text numerically for similarity comparison. Text generation is performed separately by Large Language Models (LLMs).

5. Retrieval Quality Depends on Embeddings

Poor embeddings can lead to:

irrelevant retrieval
weak semantic matching
hallucinated answers in RAG systems

This highlights the importance of selecting the correct embedding model and validating similarity scores.

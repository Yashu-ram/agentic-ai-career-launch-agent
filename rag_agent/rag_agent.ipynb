Retrieval Score Logging

The RAG system logs retrieval distance scores alongside retrieved chunks. These scores help evaluate semantic relevance between the user query and stored document embeddings.

Example:

Distance Score: 1.47

In ChromaDB, lower distance scores indicate higher semantic similarity. Logging retrieval scores helps debug retrieval quality before modifying prompts or LLM behavior.

This is important because many RAG failures are retrieval failures rather than generation failures.


2) Retrieval Failure Example:


Question

What is the CEO salary?
Observation

The retrieved chunks were not directly related to the question. The system retrieved semantically weak or unrelated chunks because the required information did not exist in the document collection.

Reason for Failure

The failure occurred because:

the answer was not present in the indexed documents
embeddings still attempted to find semantically nearest chunks
vector retrieval always returns nearest vectors even if relevance is weak
Key Learning

RAG systems can fail when:

documents lack required information
chunking is poor
embeddings are too general
retrieval thresholds are not enforced

This demonstrated the importance of:

inspecting retrieval scores
validating chunk relevance
implementing confidence thresholds
preventing hallucinations
